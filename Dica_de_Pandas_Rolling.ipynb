```python
import pandas as pd
import numpy as np
import random
```

```python
data = []
date_range = pd.date_range("01/01/2022", "01/07/2023")

for i in range(len(date_range)):

    sales_cash = random.randint(1, 10)
    sales_card = random.randint(1, 10)

    day_type = "Dia da Semana"

    if date_range[i].weekday()>4:
        day_type = "Final de Semana"

    data.append([str(date_range[i].date()), day_type, "Dinheiro", sales_cash])
    data.append([str(date_range[i].date()), day_type, "Cartão", sales_card])

data = pd.DataFrame(data, columns=["Data", "Tipo_Dia", "Tipo_Pagamento", "Vendas"])
data["Data"] = pd.to_datetime(data["Data"])
```

```python
data
```

```python
data['roll_soma_win2'] = data.rolling(window=2).Vendas.sum()
```

```python
data['roll_avg_win3'] = data.rolling(window=3).Vendas.mean()
```

```python
data.drop('roll_max_win2d', axis=1, inplace=True)
```

```python
data['roll_max_win2'] = data.rolling(window=2).Vendas.max()
```

```python
data['roll_soma_win2d'] = data.rolling(window='7d', on='Data').Vendas.sum()
```
